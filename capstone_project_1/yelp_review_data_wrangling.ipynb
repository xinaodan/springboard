# Data
Yelp open dataset is used for the project (https://www.yelp.com/dataset).
The data is pretty clean.

The purpose of the wrangling is to get the reviews of Toronto restaurant from the dataset.

# Related Files
review.json

```python
import pandas as pd
reviews = pd.read_json('review_sample.json', lines=True)
reviews.head()
```

business.json

```python
business = pd.read_json('business.json', lines=True)
business.head()
```

# Steps
1. Get Toronto restaurant from business.json by categories and city.

```python
toronto_business = business[business.city == 'Toronto']
toronto_restaurant = toronto_business[toronto_business['categories'].str.contains('Restaurants', na=True)]
toronto_restaurant.head()
```

2. Read reviews from review.json by chunk and merge it with Toronto restaurants.

```python
%%time

reader = pd.read_json('review.json', chunksize=100000, lines=True)
toronto_restaurant_review = pd.DataFrame()
for chunk in reader:
    toronto_restaurant_review = toronto_restaurant_review.append(chunk.merge(toronto_restaurant, on='business_id'))
```

```python
toronto_restaurant_review.head()
```

3. Reset business_id and review_id as indexes.

```python
toronto_restaurant_review = toronto_restaurant_review.set_index(['business_id', 'review_id'])
toronto_restaurant_review.head()
```

4. Rename columns

```python
toronto_restaurant_review = toronto_restaurant_review.rename(
    columns={'stars_x': 'stars_business', 'stars_y': 'stars_review'})
toronto_restaurant_review.head()
```
